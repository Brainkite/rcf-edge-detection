# Metrics

## Crisp Loss (inversed Dice score + BCEL)

```python
def inv_dice_loss(pred, targ):
    eps = 1e-7
    p = pred.view(-1)
    t = targ.view(-1)
    res = (p**2).sum() + (t**2).sum() + eps
    res /= 2 * (p*t).sum() + eps
    return res
```

```python
def crisp_loss(pred,targ, a=1., b=.001):
    pred = pred[-1]
    CE_loss = F.binary_cross_entropy(pred, targ)
    D_loss = inv_dice_loss(pred,targ)
    loss = a * D_loss + b * CE_loss
    return loss
```

## F1_Score

```python
def get_precision(pred,targ):
    eps = 1e-7
    tp = (pred*targ).sum()
    fp = (pred*(1-targ)).sum()
    return tp / (tp + fp + eps)
```

```python
def get_recall(pred,targ):
    eps = 1e-7
    tp = (pred*targ).sum()
    fn = ((1-pred)*targ).sum()
    return tp / (tp + fn + eps)
```

```python
def f1_score(pred,targ):
    eps = 1e-7
    precision = get_precision(pred,targ)
    recall = get_recall(pred,targ)
    
    f1 = (2*precision*recall)/(precision+recall+eps)
    return f1
```

## Quantitative Scores

### Binary pred

```python
def binary_pred(pred, thresh=.5):
    b = pred.clone().detach()
    b[b < thresh] = 0
    b[b >= thresh] = 1
    return b
```

### ODS

```python
def ods(preds, targ, return_all=False):
    p = preds[-1]
    scores = { f1_score(binary_pred(p,thresh=t), targ) : t  for t in np.linspace(0.,1.,num=100)}
    best_score = max(scores.keys())
    best_thresh = scores[best_score]
    if return_all: return best_score, best_thresh, scores
    else: return best_score
```

# test-pred CB

```python
class PredEvo(Callback):
    '''
    Callback that produces and saves a set of image predictions after each batch
    '''
    def __init__(self, tdl, out_dir):
        self.tdl = tdl
        self.out_dir = out_dir
        self.count = 0
        
    def after_batch(self):
        print('CBBB')
        self.count += 1
        
        if self.count % 2 == 0:
            batch = self.tdl.one_batch()
            y = self.model(batch[0])
            p = y[-1]
            #p = torch.sigmoid(p)
            n = p.shape[0]
            fig, axes = plt.subplots(nrows=1, ncols=n, figsize=(n*5, 5))
            for i,ax in zip(range(n), axes.flatten()):
                ax.imshow(p[i,0].detach().cpu().numpy(), cmap='gray')
            fig.suptitle('Epoch: '+str(self.epoch), fontsize=16)
            fig.tight_layout()
            fig.savefig(str(self.out_dir/str(self.epoch))+'.jpg', dpi=150, format='jpg')
```
